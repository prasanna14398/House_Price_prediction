```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Chennai houseing sale.csv")
df.head()
```

```python
df.shape
```

```python
df['N_BEDROOM'].fillna(df['N_BEDROOM'].mean(),inplace=True)
df['N_BATHROOM'].fillna(df['N_BATHROOM'].mean(),inplace=True)
df['QS_OVERALL'].fillna(df['QS_OVERALL'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.info()
```

```python
df_correl=df.drop('PRT_ID',axis='columns').corr(numeric_only=True)
plt.figure(figsize=(10,8))
sns.heatmap(df_correl,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
df.columns
```

```python
input=df.drop(['PRT_ID','DIST_MAINROAD','DATE_SALE','SALE_COND','QS_ROOMS','QS_BATHROOM','DATE_BUILD','BUILDTYPE','UTILITY_AVAIL','MZZONE'],axis='columns')
input.head()
```

```python
categorical_col=['AREA','PARK_FACIL','STREET']
```

```python
df_encoded=pd.get_dummies(input,columns=categorical_col,drop_first=True)
df_encoded.head()
```

```python
df_encoded[df_encoded.select_dtypes('bool').columns]=df_encoded.select_dtypes('bool').astype(int)
```

```python
df_encoded.info()
```

```python
import scipy.stats as stats
z_score=stats.zscore(df_encoded)
threshold=3
print("size before removing outliers",df_encoded.shape)
outliers_df=df_encoded[(abs(z_score)>threshold).any(axis=1)]
df_encoded=df_encoded[(abs(z_score)<threshold).all(axis=1)]
print("size after removing outliers",df_encoded.shape)
```

```python
x=df_encoded.drop('SALES_PRICE',axis='columns')
y=df_encoded['SALES_PRICE']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_encoded=sc.fit_transform(x_train)
x_test_encoded=sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train_encoded,y_train)
```

```python
lr.score(x_train_encoded,y_train)
```

```python
lr.score(x_test_encoded,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
lg.fit(x_train_encoded,y_train)
```

```python
lg.score(x_train_encoded,y_train)
```

```python
lg.score(x_test_encoded,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(x_train_encoded,y_train)
```

```python
dt.score(x_train_encoded,y_train)
```

```python
dt.score(x_test_encoded,y_test)
```

```python
from sklearn.svm import SVR
sv=SVR(kernel='linear')
sv.fit(x_train_encoded,y_train)
```

```python
sv.score(x_train_encoded,y_train)
```

```python
sv.score(x_train_encoded,y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, accuracy_score, precision_score, recall_score, f1_score

# Linear Regression
y_pred_lr = lr.predict(x_test_encoded)
mae_lr = mean_absolute_error(y_test, y_pred_lr)
mse_lr = mean_squared_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)

print("Linear Regression Metrics:")
print(f"Mean Absolute Error (MAE): {mae_lr}")
print(f"Mean Squared Error (MSE): {mse_lr}")
print(f"R-squared (R2): {r2_lr}")
print("-" * 30)

# Decision Tree Regressor
y_pred_dt = dt.predict(x_test_encoded)
mae_dt = mean_absolute_error(y_test, y_pred_dt)
mse_dt = mean_squared_error(y_test, y_pred_dt)
r2_dt = r2_score(y_test, y_pred_dt)

print("Decision Tree Regressor Metrics:")
print(f"Mean Absolute Error (MAE): {mae_dt}")
print(f"Mean Squared Error (MSE): {mse_dt}")
print(f"R-squared (R2): {r2_dt}")
print("-" * 30)

# Support Vector Regressor
y_pred_sv = sv.predict(x_test_encoded)
mae_sv = mean_absolute_error(y_test, y_pred_sv)
mse_sv = mean_squared_error(y_test, y_pred_sv)
r2_sv = r2_score(y_test, y_pred_sv)

print("Support Vector Regressor Metrics:")
print(f"Mean Absolute Error (MAE): {mae_sv}")
print(f"Mean Squared Error (MSE): {mse_sv}")
print(f"R-squared (R2): {r2_sv}")
print("-" * 30)
```
